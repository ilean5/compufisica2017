# Crecimiento Logistico.
La función logística,es una función matemática que aparece en diversos modelos de crecimiento de poblaciones, propagación de enfermedades epidémicas y difusión en redes sociales. Dicha función constituye un refinamiento del modelo exponencial para el crecimiento de una magnitud. Modela la función sigmoidea de crecimiento de un conjunto P.
Esta típica aplicación de la ecuación logística es un modelo común del crecimiento poblacional según el cual: 
•	La tasa de reproducción es proporcional a la población existente, es decir, mientras la población no sufra limitaciones de los recursos esta puede crecer exponencialmente y por lo tanto, puede modelarse por: dN/dt= r^N 
•	La tasa de reproducción es proporcional a la cantidad de recursos disponibles, es decir, la población deja de crecer cuando ha alcanzado la capacidad de carga del sistema, que es igual a k individuos. 


# Ejemplo 1. 
Enuentra la grafica que describe la funcion logistica considerando una poblacion incial de 100 individuos, con una tasa de crecimiento del 2% y una capacidad de carga de 400 individuos. Se define a la funcion logistica como: (p_0*k)/(p_0+((k-p_0)*e^(-r*t)))


```julia
using Plots 
```

```julia
p_0=100.0 #suponemos una poblacion incial de 100
```

```julia
r=0.2 #tasa de crecimiento del 2%
```

```julia
t = collect(0:100); # Definimos al tiempo de 0 a 100
```

```julia
e=2.7182818284590452
```

```julia
k=200.0 #Consideramos una capacidad de carga de 400 y 200
```

```julia
P = zeros(t); #Definimos a P a partir de cero 
```

```julia
for i in eachindex(t)
    P[i] = (Float64((p_0*k)))/(Float64(p_0+((k-p_0)*e^(-r*t[i]))))
end
#Usamos un ciclo for para hacer una regla de asignacion para indicar un valor de p para cada t. 
```

```julia
plot(t,P,w=2) # con k=400 y r=0.2 
title!("Crecimiento Logistico")
xaxis!("Tiempo (hr)")
yaxis!("Poblacion (bac)")
# Se grafica, indicando el titulo de la grafica y el nombre de los ejes. 
```

```julia
plt = plot(1,xlim=(0,500),ylim=(0,500),xlab="tiempo(hr))",ylab="poblacion (bac)",title="crecimiento logistico")
# Se ajusta la escala de los ejes y se indica el nombre de la grafica y los ejes. 
```
